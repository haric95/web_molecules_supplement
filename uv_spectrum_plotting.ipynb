```python
import csv
import matplotlib.pyplot as plt
```

```python
def plot_ir(filepath, save):
    wavenumbers = []
    intensities = []
    with open(filepath, "r") as stream:
        values = csv.reader(stream, delimiter = "\t")
        next(values, None)
        for i in values:
            wavenumbers.append(float(i[0]))
            intensities.append(float(i[1]) / 100)
    ymin = min(intensities) - 0.1
    plt.figure(figsize = (16,10))
    plt.plot(wavenumbers, intensities, color = "black")
    plt.grid(True, linewidth = 0.5, linestyle = "--")
#     vlines = [500,1000,1500,2000,2500,3000,3500]
#     for i in vlines:
#         plt.axvline(i, linestyle = "--")
    plt.xlim(min(wavenumbers),max(wavenumbers))
    plt.ylabel("Absorption Intensity", fontsize = 16)
    plt.xlabel("Wavenumber / nm", fontsize = 16)
    if save:
        plt.savefig(filepath[:-3] + "png")
```

```python
molecule = input("Molecule Name: ")
filepath = "./" + molecule + "/" + molecule + "_uv.txt"
```

```python
plot_ir(filepath, True)
```
